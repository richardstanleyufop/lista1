```python
from mip import Model, xsum, INTEGER,CONTINUOUS, MAXIMIZE, MINIMIZE
```

```python
m = Model('atv2a',sense=MAXIMIZE)
```

```python
x1=m.add_var(var_type=INTEGER)
x2=m.add_var(var_type=INTEGER)
```

```python
m+=15*x1+40*x2
```

```python
m+=x1<=4*x2
m+=x2+(x1-2*x2)/6<=680
```

```python
m.optimize()
```
